```python
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import FAISS
from langchain_upstage import ChatUpstage
from langchain.chains import RetrievalQA
from langchain.prompts import PromptTemplate
from langchain_upstage import UpstageEmbeddings

print("==> 1. ë¬¸ì„œ ë¡œë”© â†’ PDF ì½ê¸°...")
file_path = "../data/tutorial-korean.pdf"
loader = PyPDFLoader(file_path)
documents = loader.load()
print(f"  ì´ {len(documents)}í˜ì´ì§€ ë¡œë“œ ì™„ë£Œ")
```

```python

print("==> 2. ë¬¸ì„œ ë¶„í•  â†’ ì‘ì€ ì²­í¬ë¡œ ë‚˜ëˆ„ê¸°")
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,        # ì²­í¬ í¬ê¸° (í•œêµ­ì–´ ìµœì í™”)
    chunk_overlap=200,      # ì¤‘ë³µ ë¶€ë¶„ (ë§¥ë½ ë³´ì¡´)
    separators=["\n\n", "\n", ".", " ", ""] # ìì—°ìŠ¤ëŸ¬ìš´ ë¶„í• ì„ ìœ„í•œ êµ¬ë¶„ì
)

chunks = text_splitter.split_documents(documents)
print(f"  {len(chunks)}ê°œ ì²­í¬ ìƒì„± ì™„ë£Œ")
print(f"  í‰ê·  ì²­í¬ ê¸¸ì´: {sum(len(chunk.page_content) for chunk in chunks) / len(chunks):.0f}ì")
print(type(chunks[0]))
```

```python

print("==> 3. ë²¡í„°í™” â†’ ì„ë² ë”©ìœ¼ë¡œ ë³€í™˜")
embeddings = UpstageEmbeddings(model="solar-embedding-1-large")

print("==> 4. ì €ì¥ â†’ FAISS ë²¡í„°ìŠ¤í† ì–´ì— ì €ì¥")
vectorstore = FAISS.from_documents(chunks, embeddings)
print(f" FAISS ë²¡í„°ìŠ¤í† ì–´ ìƒì„± ì™„ë£Œ ({len(chunks)}ê°œ ë²¡í„°)")
# ë¡œì»¬ íŒŒì¼ë¡œ ì €ì¥
vectorstore.save_local("faiss_db")

print("===> 5. ê²€ìƒ‰ â†’ ì§ˆë¬¸ê³¼ ìœ ì‚¬í•œ ë¬¸ì„œ ì°¾ê¸°")
retriever = vectorstore.as_retriever(
    search_type="similarity",
    search_kwargs={"k": 6}  # ìƒìœ„ 6ê°œ ê´€ë ¨ ë¬¸ì„œ ê²€ìƒ‰
)
print(" Retriever ì„¤ì • ì™„ë£Œ")
```

```python

print("===> 6. ìƒì„± â†’ LLMìœ¼ë¡œ ë‹µë³€ ìƒì„±")
llm = ChatUpstage(
        model="solar-pro",
        base_url="https://api.upstage.ai/v1",
        temperature=0.5
    )
print(llm)
```

```python

# í•œêµ­ì–´ ìµœì í™” í”„ë¡¬í”„íŠ¸
prompt_template = """
ë‹¹ì‹ ì€ BlueJ í”„ë¡œê·¸ë˜ë° í™˜ê²½ ì „ë¬¸ê°€ì…ë‹ˆë‹¤. 
ì•„ë˜ ë¬¸ì„œ ë‚´ìš©ì„ ë°”íƒ•ìœ¼ë¡œ ì •í™•í•˜ê³  ì¹œì ˆí•œ ë‹µë³€ì„ ì œê³µí•´ì£¼ì„¸ìš”.

ë¬¸ì„œ ë‚´ìš©:
{context}

ì§ˆë¬¸: {question}

ë‹µë³€ ê·œì¹™:
1. ë¬¸ì„œ ë‚´ìš©ë§Œì„ ê·¼ê±°ë¡œ ë‹µë³€í•˜ì„¸ìš”
2. ë‹¨ê³„ë³„ ì„¤ëª…ì´ í•„ìš”í•˜ë©´ ìˆœì„œëŒ€ë¡œ ì‘ì„±í•˜ì„¸ìš”  
3. êµ¬ì²´ì ì¸ ë©”ë‰´ëª…, ë²„íŠ¼ëª…ì„ í¬í•¨í•˜ì„¸ìš”
4. ë¬¸ì„œì— ì—†ëŠ” ì •ë³´ëŠ” "ë¬¸ì„œì—ì„œ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤"ë¼ê³  í•˜ì„¸ìš”

ë‹µë³€:"""

prompt = PromptTemplate(
    template=prompt_template,
    input_variables=["context", "question"]
)
print(" í”„ë¡¬í”„íŠ¸ ì„¤ì • ì™„ë£Œ")
print(prompt)
```

```python

# ===================================
# 7. QA ì²´ì¸ ìƒì„±
# ===================================
print("\n ===> 7.  QA ì²´ì¸ ìƒì„±...")
# LLMê³¼ Retriever ì—°ê²°
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever,
    chain_type_kwargs={"prompt": prompt},
    return_source_documents=True
)
print("  RAG íŒŒì´í”„ë¼ì¸ êµ¬ì¶• ì™„ë£Œ!")

# ===================================
# 8. í…ŒìŠ¤íŠ¸ ì§ˆë¬¸ë“¤
# ===================================
test_questions = [
    "BlueJì—ì„œ ê°ì²´ë¥¼ ìƒì„±í•˜ëŠ” ë°©ë²•ì€ ë¬´ì—‡ì¸ê°€ìš”?",
    "ì»´íŒŒì¼ ì˜¤ë¥˜ê°€ ë°œìƒí–ˆì„ ë•Œ ì–´ë–»ê²Œ í™•ì¸í•  ìˆ˜ ìˆë‚˜ìš”?", 
    "ë””ë²„ê¹…ì„ ìœ„í•´ ì¤‘ë‹¨ì ì„ ì„¤ì •í•˜ëŠ” ë°©ë²•ì„ ì•Œë ¤ì£¼ì„¸ìš”",
    "ì½”ë“œíŒ¨ë“œëŠ” ë¬´ì—‡ì´ê³  ì–´ë–»ê²Œ ì‚¬ìš©í•˜ë‚˜ìš”?",
    "ì• í”Œë¦¿ì„ ë§Œë“¤ê³  ì‹¤í–‰í•˜ëŠ” ë°©ë²•ì„ ì„¤ëª…í•´ì£¼ì„¸ìš”"
]

print("\n" + "=" * 60)
print(" RAG ì‹œìŠ¤í…œ í…ŒìŠ¤íŠ¸")
print("=" * 60)

# ===================================
# 9. ì§ˆë¬¸ ë° ë‹µë³€ ì‹¤í–‰
# ===================================
for i, question in enumerate(test_questions, 1):
    print(f"\nã€í…ŒìŠ¤íŠ¸ {i}/5ã€‘")
    print(f" ì§ˆë¬¸: {question}")
    print(" ë‹µë³€ ìƒì„± ì¤‘...")
    
    # RAG ì‹¤í–‰
    result = qa_chain.invoke({"query": question})
    answer = result["result"]
    source_docs = result["source_documents"]
    
    print(f"\n ë‹µë³€:")
    print("-" * 50)
    print(answer)
    
    # ì°¸ì¡° ë¬¸ì„œ ì •ë³´
    print(f"\n ì°¸ì¡° ë¬¸ì„œ:")
    for j, doc in enumerate(source_docs[:3], 1):
        page = doc.metadata.get('page', 'N/A')
        preview = doc.page_content[:80].replace('\n', ' ')
        print(f"   {j}. í˜ì´ì§€ {page}: {preview}...")
    
    print("\n" + "-" * 40)
```
